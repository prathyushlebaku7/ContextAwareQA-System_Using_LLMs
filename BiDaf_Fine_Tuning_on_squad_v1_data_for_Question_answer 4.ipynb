```python
from datasets import load_dataset
import pandas as pd

# Load the dataset
dataset = load_dataset("squad_v2")

train_set = dataset["train"].select(range(5000))

# Convert to DataFrame
df = pd.DataFrame(train_set)
df
```

```python
df.isnull().sum()
```

```python
# Extract required columns
df = df[['id', 'context', 'question', 'answers']]

# Ensure 'answers' contain valid start positions and text
df = df[df['answers'].apply(lambda x: len(x['text']) > 0 and len(x['answer_start']) > 0)]

# Verify DataFrame
print(df.info())
```

```python
from sklearn.model_selection import train_test_split

# Train-test split (80% train, 20% validation)
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

print(f"Training set: {len(train_df)} samples")
print(f"Validation set: {len(val_df)} samples")
```

```python
from transformers import AutoTokenizer
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm  # For progress bars

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased", use_fast=True)

# Tokenization and Label Alignment
def tokenize_and_align_labels(example):
    tokenized_inputs = tokenizer(
        example["context"],
        example["question"],
        truncation=True,
        padding="max_length",
        max_length=512,
        return_offsets_mapping=True,
    )

    start_char = example["answers"]["answer_start"][0]
    end_char = start_char + len(example["answers"]["text"][0])
    offsets = tokenized_inputs["offset_mapping"]
    token_start, token_end = None, None

    for idx, (start, end) in enumerate(offsets):
        if start_char >= start and start_char < end:
            token_start = idx
        if end_char > start and end_char <= end:
            token_end = idx
            break

    tokenized_inputs["start_positions"] = token_start or 0
    tokenized_inputs["end_positions"] = token_end or 0
    del tokenized_inputs["offset_mapping"]
    return tokenized_inputs

# Dataset Class
class QADataset(Dataset):
    def __init__(self, df):
        self.data = df
        self.tokenized_data = [tokenize_and_align_labels(row) for _, row in df.iterrows()]

    def __len__(self):
        return len(self.tokenized_data)

    def __getitem__(self, idx):
        return {
            key: torch.tensor(val) if isinstance(val, list) else torch.tensor([val])
            for key, val in self.tokenized_data[idx].items()
        }

# Define BiDAF Model
class BiDAF(nn.Module):
    def __init__(self, embedding_dim, hidden_dim, vocab_size, dropout_prob=0.3):
        super(BiDAF, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.dropout = nn.Dropout(dropout_prob)
        self.lstm = nn.LSTM(
            embedding_dim, hidden_dim, bidirectional=True, batch_first=True
        )
        self.output_layer = nn.Linear(hidden_dim * 2, 2)

    def forward(self, input_ids):
        embedded = self.dropout(self.embedding(input_ids))
        lstm_out, _ = self.lstm(embedded)
        logits = self.output_layer(lstm_out)
        return logits

# Data Preparation
train_dataset = QADataset(train_df)
val_dataset = QADataset(val_df)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

# Initialize Model, Optimizer, and Loss
vocab_size = tokenizer.vocab_size
model = BiDAF(embedding_dim=256, hidden_dim=512, vocab_size=vocab_size).to('cuda')
optimizer = Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)
criterion = nn.CrossEntropyLoss()

# Training Loop
epochs = 20
for epoch in range(epochs):
    model.train()
    total_loss = 0
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch + 1}")

    for batch in progress_bar:
        input_ids = batch["input_ids"].squeeze(1).to('cuda')
        start_positions = batch["start_positions"].squeeze(1).to('cuda')
        end_positions = batch["end_positions"].squeeze(1).to('cuda')

        optimizer.zero_grad()
        logits = model(input_ids)
        start_logits, end_logits = logits.split(1, dim=-1)

        loss_start = criterion(start_logits.squeeze(-1), start_positions)
        loss_end = criterion(end_logits.squeeze(-1), end_positions)
        loss = (loss_start + loss_end) / 2

        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        progress_bar.set_postfix({"Loss": total_loss / (progress_bar.n + 1)})

    scheduler.step(total_loss)
    print(f"Epoch {epoch + 1}, Loss: {total_loss:.4f}")


```

```python
from sklearn.metrics import f1_score

# Function to compute Exact Match and F1 score
def evaluate(model, data_loader, tokenizer):
    model.eval()
    total_em = 0
    total_f1 = 0
    n_samples = 0

    for batch in tqdm(data_loader, desc="Evaluating"):
        input_ids = batch["input_ids"].squeeze(1).to('cuda')
        start_positions = batch["start_positions"].squeeze(1).to('cuda').cpu().numpy()
        end_positions = batch["end_positions"].squeeze(1).to('cuda').cpu().numpy()

        # Model predictions
        with torch.no_grad():
            logits = model(input_ids)
            start_logits, end_logits = logits.split(1, dim=-1)
            start_logits = start_logits.squeeze(-1).cpu().numpy()
            end_logits = end_logits.squeeze(-1).cpu().numpy()

        # Predicted start and end positions
        pred_start_positions = start_logits.argmax(axis=1)
        pred_end_positions = end_logits.argmax(axis=1)

        # Compute Exact Match and F1 score for each example
        for i in range(len(start_positions)):
            true_start, true_end = start_positions[i], end_positions[i]
            pred_start, pred_end = pred_start_positions[i], pred_end_positions[i]

            # Extract predicted and true text spans
            input_tokens = tokenizer.convert_ids_to_tokens(input_ids[i].cpu().numpy())
            pred_text = tokenizer.convert_tokens_to_string(input_tokens[pred_start:pred_end + 1])
            true_text = tokenizer.convert_tokens_to_string(input_tokens[true_start:true_end + 1])

            # Compute EM
            em = 1 if pred_text.strip() == true_text.strip() else 0
            total_em += em

            # Compute F1 score
            true_tokens = set(true_text.split())
            pred_tokens = set(pred_text.split())
            common_tokens = true_tokens.intersection(pred_tokens)

            precision = len(common_tokens) / len(pred_tokens) if pred_tokens else 0
            recall = len(common_tokens) / len(true_tokens) if true_tokens else 0
            f1 = (2 * precision * recall) / (precision + recall) if precision + recall > 0 else 0
            total_f1 += f1

        n_samples += len(start_positions)

    # Calculate overall EM and F1
    em_score = total_em / n_samples * 100
    f1_score = total_f1 / n_samples * 100

    return em_score, f1_score

# Evaluate on validation set
em, f1 = evaluate(model, val_loader, tokenizer)
print("\n\n")
print(f"Exact Match (EM): {em:.2f}%")
print(f"F1 Score: {f1:.2f}%")
```

```python
# Inference Function
def infer(contexts, questions, model, tokenizer):
    model.eval()  # Set model to evaluation mode
    answers = []

    for context, question in zip(contexts, questions):
        # Tokenize input for inference
        inputs = tokenizer(
            context,
            question,
            truncation=True,
            padding="max_length",
            max_length=512,
            return_tensors="pt"
        ).to('cuda')

        # Get model predictions
        with torch.no_grad():
            logits = model(inputs["input_ids"])
            start_logits, end_logits = logits.split(1, dim=-1)
            start_logits = start_logits.squeeze(-1)
            end_logits = end_logits.squeeze(-1)

            # Identify start and end token positions
            start_index = torch.argmax(start_logits, dim=1).item()
            end_index = torch.argmax(end_logits, dim=1).item()

        # Decode the predicted answer
        predicted_tokens = inputs["input_ids"][0][start_index:end_index + 1]
        answer = tokenizer.decode(predicted_tokens, skip_special_tokens=True).strip()
        answers.append(answer)

    return answers

# Sample Contexts and Questions
contexts = [
    "The Eiffel Tower is one of the most famous landmarks in Paris. It was built in 1889 and stands over 300 meters tall.",
    "The Great Wall of China stretches across northern China and was built over several centuries to protect against invasions.",
    "Albert Einstein developed the theory of relativity, which has had a significant impact on modern physics and cosmology."
]

questions = [
    "When was the Eiffel Tower built?",
    "What was the purpose of the Great Wall of China?",
    "Who developed the theory of relativity?"
]

# Run Inference
predicted_answers = infer(contexts, questions, model, tokenizer)

# Print Results
for idx, (question, answer) in enumerate(zip(questions, predicted_answers)):
    print(f"Question {idx + 1}: {question}")
    print(f"Answer: {answer}")
    print("-" * 50)
```
